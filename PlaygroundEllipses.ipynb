```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
y = lambda x, a: (b(a)/a)*np.sqrt(a**2 - (x-np.sqrt(a**2-b(a)**2))**2) # 
```

```python
period = 1
N = np.arange(2,10)
epsilon = 2/3
b = lambda a: a * np.sqrt(1 - epsilon**2)
```

```python
plt.figure(figsize=(10, 4))

for i in N:
    xs = np.linspace(-i*period, i*period, 1000)
    plt.plot(xs, y(xs, i*period))
    plt.scatter(epsilon*i*period-np.sqrt((i*period)**2-b(i*period)**2),0)
    
plt.xlim(-10,10)
plt.ylim(0,8)
plt.show()
```
