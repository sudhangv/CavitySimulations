```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x0 = np.load('opt_result_x0.npy')
fval = np.load('opt_result_fval.npy')
grid = np.load('opt_result_grid.npy')
Jout = np.load('opt_result_Jout.npy')
```

```python
for i, dc in enumerate(grid[2,0,0,:]):

    plt.figure(figsize=(8,8))
    plt.imshow(-Jout[:,:,i], cmap='magma', vmin=0, vmax=0.8, interpolation='None')

    plt.xticks(range(grid.shape[2]), np.around(grid[1,0,:,0], 4))
    plt.yticks(range(grid.shape[1]), np.around(grid[0,:,0,0], 4))

    plt.title('Duty cycle: {}'.format(np.around(grid[2,0,0,i], 4)))

    plt.xlabel('Lattice constant [um]')
    plt.ylabel('Eccentricity')

    plt.colorbar()

    plt.show()
```
